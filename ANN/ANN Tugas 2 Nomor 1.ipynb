```python
import numpy as np
from matplotlib import pyplot
```

### No 1 (b)

```python
def objective_function(x):
    # return x**2
    return x**4-5*x**3+x**2+5*x+1
    
def derivative_function(x):
    # return 2*x
    return 4*x**3-15*x**2+2*x+5

def gradient_descent(
    gradient, start, learn_rate, n_iter=500, tolerance=1e-09
):
    cur = start
    x, y = list(), list()
    for i in range(n_iter):
        diff = -learn_rate * derivative_function(cur)
        x.append(cur)
        y.append(objective_function(cur))
        print('>iterasi %d: x = %.9f, y = %.9f' % (i + 1, cur, objective_function(cur)))
        if np.all(np.abs(diff) <= tolerance):
            break
        cur += diff
    return x, y

res_x, res_y = gradient_descent(
  gradient=derivative_function, start=4.0, learn_rate=0.05, n_iter=500
)
```

```python
ranges = np.asarray([[-2, 6]])
inputs = np.arange(ranges[0,0], ranges[0,1] + 0.1, 0.1)
results = objective_function(inputs)
pyplot.plot(inputs, results)
pyplot.plot(res_x, res_y, '.-', color='red')
pyplot.show()
```
