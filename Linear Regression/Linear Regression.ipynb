### Simple Linear Regression

```python
import numpy as np
import pandas as pd
import scipy
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
```

```python
sns.set()
```

**Example 1**<br>
SAT and GPA data

```python
data1 = pd.read_csv("sat_gpa.csv")
```

```python
data1.head(10)
```

```python
data1.describe()
```

```python
data1.dtypes
```

**Regression**<br>
y = b0 + b1x1


```python
y = data1['GPA']
x1 = data1['SAT']
```

```python
plt.scatter(x1,y)
plt.xlabel('SAT')
plt.ylabel('GPA')
plt.show()
```

```python
x = sm.add_constant(x1)
results = sm.OLS(y,x).fit()
results.summary()
```

```python
plt.scatter(x1,y)
y = 0.0017*x1 + 0.275
fig = plt.plot(x1, y, color='red', label='regression line')
plt.xlabel('SAT')
plt.ylabel('GPA')
plt.show
```

**Example 2**<br>
Real Estate Price and Size

```python
data2 = pd.read_csv("real_estate.csv")
```

```python
data2.head(10)
```

```python
y = data2['price']
x1 = data2['size']
x = sm.add_constant(x1)
results = sm.OLS(y,x).fit()
results.summary()
```

```python
plt.scatter(x1,y)
y = 223.1787*x1 + 1.019e+05
fig = plt.plot(x1, y, color='red', label='regression line')
plt.xlabel('Size')
plt.ylabel('Price')
plt.show
```

### Multiple Linear Regression

```python
data3 = pd.read_csv("real_estate_multiple.csv")
```

```python
data3.head(10)
```

```python
y = data3['price']
x1 = data3[['size', 'year']]
x = sm.add_constant(x1)
results = sm.OLS(y,x).fit()
results.summary()
```
